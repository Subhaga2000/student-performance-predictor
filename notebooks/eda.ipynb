```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/ASUS/Documents/student-performance-predictor/data/student-mat.csv", sep=';')# UCI uses semicolon
print("shape:", df.shape)
print(df.head())
print(df.info())
print(df.describe(include='all'))
```

```python
# Create binary target 'passed' (1 if final grade >=10)
df['passed'] = (df['G3'] >= 10).astype(int)
print("Passed value counts:\n", df['passed'].value_counts())
```

```python
# Missing values
print("Missing values per column:\n", df.isnull().sum())
```

```python
# Examples of plots (run in notebook)
plt.figure(figsize=(6,4))
sns.countplot(x='passed', data=df)
plt.title('Pass / Fail counts')
plt.show()
```

```python
# Correlation heatmap (numeric)
num_cols = df.select_dtypes(include=['int64','float64']).columns
plt.figure(figsize=(12,10))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Numeric Correlations')
plt.show()
```
